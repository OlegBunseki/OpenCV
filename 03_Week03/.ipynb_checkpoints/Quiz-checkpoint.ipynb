### Question 13

```python
img1 = cv2.imread('data/images/img_1.png')
img2 = cv2.imread('data/images/img_2.png')
print(img1.shape)
print(img2.shape)
plt.imshow(img1)
plt.show()
plt.imshow(img2)
plt.show()
```

```python
adj = 21
params = cv2.SimpleBlobDetector_Params()

print('blobColor:'.ljust(adj), params.blobColor)
print('filterByArea:'.ljust(adj), params.filterByArea)
print('filterByCircularity:'.ljust(adj), params.filterByCircularity)
print('filterByColor:'.ljust(adj), params.filterByColor)
print('filterByConvexity:'.ljust(adj), params.filterByConvexity)
print('filterByInertia:'.ljust(adj), params.filterByInertia)
print('maxArea:'.ljust(adj), params.maxArea)
print('maxCircularity:'.ljust(adj), params.maxCircularity)
print('maxConvexity:'.ljust(adj), params.maxConvexity)
print('maxInertiaRatio:'.ljust(adj), params.maxInertiaRatio)
print('maxThreshold:'.ljust(adj), params.maxThreshold)
print('minArea:'.ljust(adj), params.minArea)
print('minCircularity:'.ljust(adj), params.minCircularity)
print('minConvexity:'.ljust(adj), params.minConvexity)
print('minDistBetweenBlobs:'.ljust(adj), params.minDistBetweenBlobs)
print('minInertiaRatio:'.ljust(adj), params.minInertiaRatio)
print('minRepeatability:'.ljust(adj), params.minRepeatability)
print('minThreshold:'.ljust(adj), params.minThreshold)
print('thresholdStep:'.ljust(adj), params.thresholdStep)

params.filterByArea = False
print()
print('filterByArea:'.ljust(adj), params.filterByArea)

detector = cv2.SimpleBlobDetector_create(params)
```

```python
keypoints1 = detector.detect(img1)
keypoints2 = detector.detect(img2)
print(len(keypoints1))
print(len(keypoints2))
```

```python
# Mark blobs using image annotation concepts we have studied so far
im = img1.copy()

for k in keypoints1:
    x,y = k.pt
    x=int(round(x))
    y=int(round(y))
    
    # Mark center in BLACK
    cv2.circle(im,(x,y),5,(0,0,0),-1)
    
    # Get radius of blob
    diameter = k.size
    radius = int(round(diameter/2))
    
    # Mark blob in RED
    cv2.circle(im,(x,y),radius,(0,0,255),2)
    
# Let's see what image we are dealing with
plt.imshow(im[:,:,::-1])
plt.show()
```

```python
# Mark blobs using image annotation concepts we have studied so far
im = img1.copy()

for k in keypoints2:
    x,y = k.pt
    x=int(round(x))
    y=int(round(y))
    
    # Mark center in BLACK
    cv2.circle(im,(x,y),5,(0,0,0),-1)
    
    # Get radius of blob
    diameter = k.size
    radius = int(round(diameter/2))
    
    # Mark blob in RED
    cv2.circle(im,(x,y),radius,(0,0,255),2)
    
# Let's see what image we are dealing with
plt.imshow(im[:,:,::-1])
plt.show()
```

### Question 14

```python
img = cv2.imread('data/images/img_3.png')
print(img.shape)

plt.imshow(img[:,:,::-1])
plt.show()

img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
print(img_gray.shape)

plt.imshow(img_gray, cmap='gray')
plt.show()

_, img_gray_thresh = cv2.threshold(img_gray, 100, 255, cv2.THRESH_BINARY)
plt.imshow(img_gray_thresh, cmap='gray')
plt.show()

blobs, imLabels = cv2.connectedComponents(img_gray_thresh) # blobs includes the background
plt.imshow(imLabels, cmap='gray')
plt.show()

# Clean Noise!

(minVal, maxVal, minLoc, maxLoc) = cv2.minMaxLoc(imLabels)
imLabels = 255 * (imLabels - minVal)/(maxVal-minVal)
imLabels = np.uint8(imLabels)

imColorMap = cv2.applyColorMap(imLabels, cv2.COLORMAP_JET)
plt.imshow(imColorMap[:,:,::-1])
plt.show()
```
